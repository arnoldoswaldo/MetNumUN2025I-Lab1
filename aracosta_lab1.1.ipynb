```python
# Ejemplo de una clase simple
class Coche:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def mostrar_detalles(self):
        print(f"Este es un {self.marca} {self.modelo}.")

mi_coche = Coche("Toyota", "Corolla")
mi_coche.mostrar_detalles()
```

# My own example  15

```python
# Operaciones básicas con una Serie de Pandas
import yfinance as yf
import pandas as pd
ticker = "MSFT"
data = yf.download(ticker, start="2023-03-01", end="2023-03-31")
msft_prices = data['Close']
print(f"Precio máximo de MSFT en marzo: {msft_prices.max()}")
print(f"Precio mínimo de MSFT en marzo: {msft_prices.min()}")
print(f"Cambio porcentual diario:\n{msft_prices.pct_change().head()}")
```

# My own example  17

```python
# Agrupación de datos por año y cálculo del precio promedio
data_anual = yf.download("GOOG", start="2020-01-01", end="2024-12-31")
data_anual['Año'] = data_anual.index.year
precio_promedio_anual = data_anual.groupby('Año')['Close'].mean()
print(f"\nPrecio promedio anual de GOOG:\n{precio_promedio_anual}")
```

# My own example  18

```python
# Unión de dos DataFrames
data_aapl = yf.download("AAPL", start="2024-03-01", end="2024-03-31")['Close']
data_msft = yf.download("MSFT", start="2024-03-01", end="2024-03-31")['Close']
precios_combinados = pd.concat([data_aapl, data_msft], axis=1)
precios_combinados.columns = ['AAPL', 'MSFT']
print(f"\nPrecios combinados de AAPL y MSFT:\n{precios_combinados.head()}")
```

# My own example  19

```python
import matplotlib.pyplot as plt
import yfinance as yf

# Descargar datos
data = yf.download("TSLA", start="2024-01-01", end="2024-04-15")
volumen = data['Volume']

# Crear gráfico de barras usando pandas + matplotlib
plt.figure(figsize=(12, 6))
volumen.plot(kind='bar', label='Volumen')  # Esto maneja el eje x con fechas mejor
plt.title('Volumen de Transacciones de TSLA (2024)')
plt.xlabel('Fecha')
plt.ylabel('Volumen')
plt.grid(axis='y')
plt.legend()
plt.tight_layout()
plt.show()
```

# My own example  20

```python
import matplotlib.pyplot as plt
import yfinance as yf

# Descargar los datos de TSLA
data = yf.download("TSLA", start="2024-01-01", end="2024-04-15")

# Asegurar que no haya valores nulos
data = data.dropna()

# Extraer precios y volumen
precios = data['Close']
volumen = data['Volume']

# Convertir el índice a strings legibles (solo para el gráfico de barras)
# fechas = data.index.strftime('%Y-%m-%d')  # This line is no longer needed

# Crear la figura
plt.figure(figsize=(12, 8))

# Subplot 1: Precio
plt.subplot(2, 1, 1)
plt.plot(precios.index, precios, label='Precio de Cierre', color='blue')
plt.title('Precio de Cierre de TSLA (2024)')
plt.xlabel('Fecha')
plt.ylabel('Precio (USD)')
plt.grid(True)
plt.legend()

# Subplot 2: Volumen
plt.subplot(2, 1, 2)
# The plot function for Series only accepts one main argument which can be used as y axis. x axis is automatically set to index values.
volumen.plot(kind='bar', label='Volumen', color='orange')  # Changed this line
plt.title('Volumen de Transacciones de TSLA (2024)')
plt.xlabel('Fecha')
plt.ylabel('Volumen')
plt.grid(axis='y')
plt.legend()

# Mejorar visibilidad del eje X
plt.xticks(rotation=45, fontsize=8)

# Ajustar el diseño final
plt.tight_layout()
plt.show()
```
